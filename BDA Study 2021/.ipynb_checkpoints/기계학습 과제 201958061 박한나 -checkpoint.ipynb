```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np 
import pandas as pd 
import seaborn as sns
import sklearn
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import KNNImputer
```

```python
df = pd.read_csv('strokedata.csv')
```

```python
df = df.drop('id', axis=1)
```

```python
df.isnull().sum()
```

```python
labels = df['stroke'].value_counts(sort = True).index
sizes = df['stroke'].value_counts(sort = True)

colors = ["lightblue", "beige"]
explode = (0.05, 0)

plt.figure(figsize=(7,7))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True,startangle=90, )
plt.title('Number of Stroke in the dataset')
plt.show()
```

```python
label_encodin =LabelEncoder()
en_df = df.apply(label_encodin.fit_transform)
en_df.head()
```

```python
def plot_hist(col, bins=30, title="",xlabel="",ax=None):
    sns.distplot(col, bins=bins,ax=ax)
    ax.set_title('Histogram of {title}',fontsize=20)
    ax.set_xlabel(xlabel)
```

```python
x = en_df.drop(['stroke'], axis=1)
y = en_df['stroke']
```

```python
en_df_imputed = en_df
imputer = KNNImputer(n_neighbors=4, weights="uniform")
imputer.fit_transform(en_df_imputed)
```

```python
features=['gender', 'age', 'hypertension', 'heart_disease', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']

X , y = en_df_imputed[features], en_df_imputed['stroke']
x_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=23)
```

```python
def hist_score(score):
    models_names = ['Logistic Regression','Random Forest Classifier','SVM']

    plt.rcParams['figure.figsize']=20,8
    sns.set_style('darkgrid')
    ax = sns.barplot(x=models_names, y=score,  saturation =2.0)
    plt.xlabel('Classifier Models', fontsize = 20 )
    plt.ylabel('% of Accuracy', fontsize = 20)
    plt.title('Accuracy of different Classifier Models on test set', fontsize = 20)
    plt.xticks(fontsize = 12, horizontalalignment = 'center', rotation = 8)
    plt.yticks(fontsize = 12)
    for i in ax.patches:
        width, height = i.get_width(), i.get_height()
        x, y = i.get_xy() 
        ax.annotate(f'{round(height,2)}%', (x + width/2, y + height*1.02), ha='center', fontsize = 'x-large')
    plt.show()
```

```python
def run_exp_on_feature(x_train,y_train,x_test,y_test):

    models= [['Logistic Regression ',LogisticRegression()],
             ['Random Forest Classifier ',RandomForestClassifier()],
             ['SVM ',SVC()]]

    models_score = []
    for name,model in models:

        model = model
        model.fit(x_train,y_train)
        model_pred = model.predict(x_test)
        cm_model = confusion_matrix(y_test, model_pred)
        print(cm_model)
        models_score.append(accuracy_score(y_test,model.predict(x_test)))
        
        test_accuracy_score = accuracy_score(y_test,model.predict(x_test))
        training_accuracy_score = accuracy_score(y_train,model.predict(x_train))
        
        print(name)
        print('Test Acuuracy: ',test_accuracy_score)
        print('Training Accuracy: ',training_accuracy_score)
        print('############################################')
        
        ms = [[models_score]]
    return models_score
```

```python
trainset = run_exp_on_feature(x_train, y_train, x_test, y_test)
```

```python
hist_score(trainset)
```

```python
sub_train,sub_test,val_train,val_test = train_test_split(x_train,y_train,test_size=0.2,random_state=23)
```

```python
valset = run_exp_on_feature(sub_train, val_train, sub_test, val_test)
```

```python
hist_score(valset)
```

```python
models_names = ['Logistic Regression','Random Forest Classifier','SVM']

plt.figure(figsize=(17,7))
plt.plot(models_names, trainset)
plt.plot(models_names, valset)
plt.xlabel('Classifier Models', fontsize=20)
plt.ylabel('% of Accuracy', fontsize=20)
plt.title('Accuracy of different Classifier Models on Test set and Validation set', fontsize=20)
plt.legend(['Train Set', 'Validation Set'], fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=16)

plt.show()
```
