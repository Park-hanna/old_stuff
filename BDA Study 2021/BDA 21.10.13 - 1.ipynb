```python
x = np.linspace(0,10,100)
```

```python
x
```

```python
plt.plot(x, np.sin(x));
```

```python
ax = plt.axes()
ax.plot(x,np.sin(x))
```

# lineplot

```python
plt.plot(x, np.sin(x), color='red', linestyle='solid', label = 'sin(x)')
plt.plot(x, np.cos(x), color='g',linestyle='--', label ='cos(x)')
plt.title('sin & cos curves')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.ylim((-1.2,1.2))
plt.legend()
plt.tight_layout()
```

# histogram

```python
np.random.seed(1)
dat = np.random.randn(1000)
```

```python
plt.hist(dat)
```

```python
np.random.seed(2)
x1 = np.random.normal(0,0.8,1000)
x2 = np.random.normal(-2,1,1000)
```

```python
plt.hist(x1, alpha=0.5, bins=30)
plt.hist(x2, alpha=0.5, bins=30)
```
